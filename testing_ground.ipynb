```python
import numpy as np
matrix = np.array([[1, 2, 3], [4, 3, 6], [7, 6, 5]])

print(np.argmax(matrix, axis=0))
```

```python
def dist_to_one_hot(matrix):
    # Find the indices of the max values in each column
    max_indices = np.argmax(matrix, axis=1)

    # Create an empty matrix with the same shape
    one_hot_encoded = np.zeros_like(matrix)

    # Set the max positions to 1
    for col, row in enumerate(max_indices):
        one_hot_encoded[row, col] = 1

    return one_hot_encoded
```

```python
dist_to_one_hot(matrix)
```

```python
import numpy as np
from cluster import KMeans
from cluster import utils
from cluster.silhouette import Silhouette
from sklearn.metrics import silhouette_score


data, labels = utils.make_clusters()

kmeans_obj  = KMeans(3)
kmeans_obj.fit(data)
predicted = kmeans_obj.predict(data)

silhouette_obj = Silhouette()
scores = silhouette_obj.score(X=data, y=predicted)

# print(scores)
# print(np.random.choice([0, 1, 2], size=(data.shape[0], 1)))
foo_predict = np.random.choice([0, 1, 2], size=(data.shape[0]))
foo_scores = silhouette_obj.score(X=data, y=foo_predict)

# print(predicted)
utils.plot_multipanel(data, labels, foo_predict, foo_scores)
utils.plot_multipanel(data, labels, predicted, scores)
# print(scores)

true_scores = silhouette_score(data, predicted)

for i in range(scores):
    print(scores[i], true_scores[i])
```
